```python
import pandas as pd

df_nodes = pd.read_csv("network/Successful Network/Original Network/nodes_original.csv", sep="\t")
df_nodes
```

```python
df_nodes.iloc[:, 6:-1].max()
```

```python
import matplotlib.pyplot as plt


# Create the boxplot
plt.figure(figsize=(20, 12))
df_nodes.iloc[:, 6:-1].boxplot()


# Set title and labels
#plt.ylim(0,1.0*10**(-2))
plt.title('Boxplot of Features')
plt.xlabel('Features')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=len(df_nodes.iloc[:, 6:-1].columns), figsize=(15, 5))

# Plot boxplots for each column
for i, column in enumerate(df_nodes.iloc[:, 6:-1].columns):
    df_nodes[column].plot(kind='box', ax=axes[i])
    axes[i].set_title(column)

plt.tight_layout()
plt.show()
```
