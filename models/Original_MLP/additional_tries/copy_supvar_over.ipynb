```python
import pandas as pd

df_hsp_song_collab = pd.read_csv("HSP_song_collab_small_netmet.csv", sep=",")
df_hsp_song_collab
```

```python
columns = ["hit", "song_id", "artist1_id", 'artist2_id',
"success_rate_x", "superstar_x", "superstar_v5_x","superstar_v4_x","superstar_v3_x","superstar_v2_x","superstar_v1_x", "hits_in_past_x", 
"hits_in_past_y", "success_rate_y", "superstar_y", "superstar_v5_y","superstar_v4_y","superstar_v3_y","superstar_v2_y","superstar_v1_y"]

df_sup = pd.read_csv("data_superstar_v1_0_small.csv", sep=",")
df_sup = df_sup[columns]
```

```python
data = pd.merge(df_hsp_song_collab, df_sup, on=["song_id", "artist1_id", 'artist2_id', "hit"])
```

```python
# Find common columns
common_columns = data.columns.intersection(df_sup.columns)

# Calculate differences for each common column
differences = pd.DataFrame()

for col in common_columns:
    print(col, "  ", data.nunique()[col] - df_sup.nunique()[col])
    differences[col] = data.nunique()[col] - df_sup.nunique()[col]

print("Common columns:", common_columns.tolist())
print("Differences DataFrame:")
print(differences)
```
