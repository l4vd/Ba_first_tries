## Preprocessing:

### Artists:

```python
import pandas as pd
import ast 

df_artists = pd.read_csv("../../MusicOSet/network_clustered/nodes.csv", sep=",")
df_artists.drop(columns=["harmonicclosnesscentrality", "triangles", "modularity_class"], inplace=True)
df_artists["genres"].fillna('[]', inplace=True)
df_artists["genres"] = df_artists["genres"].apply(ast.literal_eval)
df_artists["amount_of_genres"] = df_artists["genres"].apply(lambda x: len(x))
df_artists
```

```python
df_releases = pd.read_csv("../../MusicOSet/musicoset_metadata/releases.csv", sep="\t")
df_releases
```

```python
# Group by 'id' and count occurrences of 'albumid'
grouped = df_releases.groupby("artist_id").agg({"album_id": "count"}).reset_index()

# Sort by the count of 'albumid' in descending order
sorted_df = grouped.sort_values(by="album_id", ascending=False)
sorted_df
```

```python
counts = {}
allready= {}
unique_artists = df_releases["artist_id"].unique()
for i in unique_artists:
    allready[i] = []

for _, row in df_releases.iterrows():
    artist_id = row["artist_id"]
    if len(allready[artist_id]) == 0:
        allready[artist_id].append(row["album_id"])
        counts[artist_id] = counts.get(artist_id, 0) + 1
    elif row["album_id"] not in allready[artist_id]:
        allready[artist_id].append(row["album_id"]) 
        counts[artist_id] = counts.get(artist_id, 0) + 1

counts
```

```python
sorted_counts = dict(sorted(counts.items(), key=lambda item: item[1], reverse=True))
sorted_counts
```

```python
# Rename the "artist_id" column to "spotify_id"
grouped.rename(columns={"artist_id": "Spotify ID", "album_id": "amount_albums"}, inplace=True)

grouped
```

```python
merged_artists_df = pd.merge(df_artists, grouped, how="left")
merged_artists_df.fillna(0, inplace=True)
merged_artists_df
```
