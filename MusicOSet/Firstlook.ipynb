```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hits_dataset.csv", delimiter="\t")
df
```

```python
df["song_id"].nunique()
```

```python
df["id_artists"].nunique()
```

```python
# Extract the release year from the "release_date" column
df['release_year'] = pd.to_datetime(df['release_date']).dt.year

# Count the number of songs per release year
songs_per_year = df['release_year'].value_counts().sort_index()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
songs_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Songs per Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Songs')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
